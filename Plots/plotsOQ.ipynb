```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# HAIR PLOT 
metrics_csv = "../misc_data/result-OQ-smote.csv"
df = pd.read_csv(metrics_csv)

# gather F1 / AUC into long form
f1_cols  = [c for c in df.columns if c.startswith("F1_fold")]
auc_cols = [c for c in df.columns if c.startswith("AUC_fold")]

long_f1  = df.melt(id_vars="Model", value_vars=f1_cols,
                   var_name="Fold", value_name="Value")
long_f1["Metric"] = "F1"

long_auc = df.melt(id_vars="Model", value_vars=auc_cols,
                   var_name="Fold", value_name="Value")
long_auc["Metric"] = "AUC"

long_df = pd.concat([long_f1, long_auc])

# colors to use
box_color = "salmon"
median_color = "black"

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=False)

for ax, metric in zip(axes, ["AUC", "F1"]):
    data = long_df[long_df["Metric"] == metric]
    bp = data.boxplot(
        column="Value",
        by="Model",
        ax=ax,
        patch_artist=True,               # <-- allow color fill
        boxprops=dict(facecolor=box_color, color=box_color),
        medianprops=dict(color=median_color),
        whiskerprops=dict(color=box_color),
        capprops=dict(color=box_color),
        flierprops=dict(markeredgecolor=box_color)
    )
    ax.set_title(f"{metric} across folds")
    ax.set_xlabel("Model")
    ax.set_ylabel(metric)
    ax.tick_params(axis="x", rotation=90)

plt.suptitle("Box-plots of AUC and F1 per fold")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File paths
files = {
    "SMOTE": "../result/results_OQ_smote_metrics.csv",
    "undersampling" : "../result/results_OQ_under_metrics.csv",
    "smote+under": "../result/results_OQ_smote+under_metrics.csv",
    "image-aug": "../result/results_OQ_augmentation_metrics.csv"
    
}

# Metrics to plot
metrics_to_plot = ["AUC", "F1", "Precision", "Recall", "Accuracy"]

# Color mapping using hex codes
colors = {
    "undersampling": "#6CA0DC",   # soft blue
    "smote+under": "#F4A582",   # soft salmon
    "SMOTE": "#C4A1E0" ,     # soft purple
    "image-aug": "#A1D6D6"
}

# Load and extract data
data = {}
for label, path in files.items():
    df = pd.read_csv(path)
    data[label] = df[metrics_to_plot].iloc[0]

plot_df = pd.DataFrame(data, index=metrics_to_plot)

# Plot
x = np.arange(len(metrics_to_plot))
width = 0.2
fig, ax = plt.subplots(figsize=(10, 6))

for i, (model_name, scores) in enumerate(plot_df.items()):
    ax.bar(x + i*width - width, scores, width, label=model_name, color=colors[model_name])

# Labels and formatting
ax.set_ylabel('Score')
ax.set_title('Comparison of data augmentated models')
ax.set_xticks(x)
ax.set_xticklabels(metrics_to_plot)
ax.set_ylim(0, 1)
ax.legend()

# Annotate bars
for i, (model_name, scores) in enumerate(plot_df.items()):
    for j, score in enumerate(scores):
        ax.text(x[j] + i*width - width, score + 0.01, f'{score:.2f}', 
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
