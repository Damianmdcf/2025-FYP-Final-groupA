```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Read the CSV file
metrics_csv = Path("..") / "data" / "baseline-data-for-model.csv"
df = pd.read_csv(metrics_csv)


# Define the features to plot
features = ['feature_a', 'feature_b', 'feature_c']

# Loop through pairs of features to make scatterplots
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        x = features[i]
        y = features[j]

        # Create a scatterplot
        plt.figure()
        for label, color in zip([0, 1], ['green', 'red']):
            subset = df[df['Melanoma'] == label]
            plt.scatter(subset[x], subset[y], label=f"Melanoma: {label}", alpha=0.7, c=color)

        plt.xlabel(x)
        plt.ylabel(y)
        plt.title(f"{x} vs {y}")
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = 'Z_feature_a'
y = 'Z_feature_c'

# Create scatterplot
plt.figure()
for label, color in zip([0, 1], ['steelblue', 'salmon']):
    subset = df[df['Melanoma'] == label]
    plt.scatter(subset[x], subset[y], label=f'Melanoma: {label}', color=color, alpha=0.5)

    # Plot group means as lines
    mean_x = subset[x].mean()
    mean_y = subset[y].mean()
    plt.axvline(mean_x, color=color, linestyle='--', alpha=1)
    plt.axhline(mean_y, color=color, linestyle='--', alpha=1)

plt.xlabel("Asymmetry score")
plt.ylabel("Color irregularity score")
plt.title("Asymmetry vs Color")
plt.legend()
plt.grid(False)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Rename columns if needed to match the example
df['label'] = df['Melanoma'].map({1: 'Melanoma', 0: 'Non-Melanoma'})

# Create the jointplot
g = sns.jointplot(
    data=df,
    x='feature_a',   # symmetry
    y='feature_c',   # color irregularity
    hue='label',
    kind='scatter',
    palette={'Melanoma': 'red', 'Non-Melanoma': 'green'},
    marginal_kws=dict(common_norm=False, fill=True, alpha=0.5),
    height=7
)

# Optional: adjust legend
g.ax_joint.legend(title='', loc='upper left')

plt.suptitle("Symmetry and Color", fontsize=14, style='italic')
plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# List your features
features = ['feature_a', 'feature_b', 'feature_c']

# Plot KDEs for each feature
plt.figure(figsize=(10, 6))

for feature in features:
    sns.kdeplot(df[feature], label=feature, fill=True, alpha=0.4, linewidth=2)

plt.title("Density Comparison of Features A, B, and C")
plt.xlabel("Feature Value")
plt.ylabel("Density")
plt.legend(title="Feature")
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: map labels for clarity
df['label'] = df['Melanoma'].map({0: 'Non-Melanoma', 1: 'Melanoma'})

# Melt the DataFrame to long format for seaborn
melted = pd.melt(df, id_vars='label', value_vars=['Z_feature_a', 'Z_feature_b', 'Z_feature_c'],
                 var_name='Feature', value_name='Value')

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=melted, x='Feature', y='Value', hue='label')

plt.title("Boxplot Comparison of Features by Class")
plt.xlabel("Feature")
plt.ylabel("Value")
plt.legend(title="Class")
plt.tight_layout()
plt.show()
```
